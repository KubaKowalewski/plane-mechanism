***Create Images of All Types of PL Linkages***

```python
from src.Slm import slm
import numpy as np

# Store link lengths for each class type [class, group, a, b, c, (scale factor for y/z group)]
a = .1
bv1,bv2,bv3 = .125,.1,.3
cv1,cv2,cv3 = .3,.2,.125
r = 0.5
v1x = [1,"x",a,bv1,cv1,1]
v2x = [2,"x",a,bv2,cv2,1]
v3x = [3,"x",a,bv3,cv3,1]
v1y = [1,"y",a,bv1,cv1,r]
v2y = [2,"y",a,bv2,cv2,r]
v3y = [3,"y",a,bv3,cv3,r]
v1z = [1,"z",a,bv1,cv1,r]
v2z = [2,"z",a,bv2,cv2,r]
v3z = [3,"z",a,bv3,cv3,r]
mechanisms = [v1x,v2x,v3x,v1y,v2y,v3y,v1z,v2z,v3z]
for mechanism in mechanisms:
    v = mechanism[0]
    g = mechanism[1]
    a,b,c = mechanism[2],mechanism[3],mechanism[4]
    r = mechanism[5]
    A = np.array([a,a])
    B = np.array([b,b,b,b])
    C = np.array([c,c])
    SLM = slm.mechanism(A,B,C,v,g,r)
    pic_path = "./images/v" + str(v) + g + ".jpg"
    SLM.draw(save=True,path=pic_path)
    # Stiffness Parmaters
    F = [0,-1000,0,0,0,0,0,0]
    N = SLM.calculate_state(SLM.theta)
    thetas = SLM.find_link_angles(N)
    E = 200e9
    D = 5
    # SLM.draw()
    SLM.path_stiffness(F,thetas,E,D)
    pic_path = "./images/k" + str(v) + g + ".jpg"
    SLM.plot_stiffness(save=True,path=pic_path)
    
```
